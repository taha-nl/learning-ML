```python
#### Splitting data
X=df.drop(["id","diagnosis"],axis=1)
y=df["diagnosis"]
```

```python
#### splitting data to train-test data

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=42,stratify=y)
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
#### decision tree

from sklearn.tree import DecisionTreeClassifier

dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
pred_test=dt.predict(X_test)
pred_train=dt.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
test_score=accuracy_score(y_test,pred_test)
train_score=accuracy_score(y_train,pred_train)
```

```python
print("train score: ",train_score)
print("test score: ",test_score)
```

**observation**: we have overfitting

```python
### confusion matrix 
from sklearn.metrics import confusion_matrix,classification_report

print(confusion_matrix(y_test,pred_test))
print(classification_report(y_test,pred_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#### optimimizing model
param_grid = {
    'max_depth': list(range(3,12,2)),
    'min_samples_split': list(range(3,10)),
    'min_samples_leaf': list(range(1,10))
}

# Create an instance of the DecisionTreeClassifier
dt = DecisionTreeClassifier()

# Create an instance of the GridSearchCV class
grid_search = GridSearchCV(dt, param_grid, cv=5)

# Fit the GridSearchCV class to the training data
grid_search.fit(X_train, y_train)

# Select the best combination of hyperparameters
best_params = grid_search.best_params_

# Train the model with the best combination of hyperparameters
best_dt = DecisionTreeClassifier(**best_params)
best_dt.fit(X_train, y_train)

# Evaluate the performance of the model on the test set
accuracy_test = best_dt.score(X_test, y_test)
accuracy_train = best_dt.score(X_train, y_train)
print("Accuracy test:", accuracy_test)
print("Accuracy train:", accuracy_train)

```
